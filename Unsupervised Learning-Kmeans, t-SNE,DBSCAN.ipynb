```python
# Importing modules
from sklearn import datasets
from sklearn.cluster import  KMeans
```

```python
# Loading Dataset
iris_df = datasets.load_iris()
```

```python
from sklearn import metrics
from scipy.spatial.distance import cdist
```

```python
## K means determine
distortions = []
K = range(1,10)
print('Iris Data Size:',iris_df.data.shape[0])
for k in K:
#     print(k)
    kmean_model = KMeans(n_clusters=k).fit(iris_df.data)
    kmean_model.transform(iris_df.data)
#     kmean_model.fit(iris_df.data)
    print('n_clusters= ',k,'Wthin-Cluser Sum of Square:',sum(np.min(cdist(iris_df.data, kmean_model.cluster_centers_, 'euclidean'), axis=1))/iris_df.data.shape[0])
    distortions.append(sum(np.min(cdist(iris_df.data, kmean_model.cluster_centers_, 'euclidean'), axis=1)) / iris_df.data.shape[0])
    
```

```python
plt.plot(K, distortions, 'g', marker = '*')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

So it is clear from above picture. * k = 3*

### Silhouette Coeficient Method
http://scikit-learn.org/stable/modules/clustering.html#silhouette-coefficient

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coeff = []
K = range(2,10)
for k in K:
    kmeans_model = KMeans(n_clusters=k).fit(iris_df.data)
    kmeans_model.transform(iris_df.data)
    label = kmeans_model.labels_
    sil_coeff = silhouette_score(iris_df.data, label, metric='euclidean')
    print('for n_clusters ={}, Silhouette coeff is: {}'.format(k,sil_coeff))
    silhouette_coeff.append(sil_coeff)
    
```

```python
kmeans_model.labels_
```

```python
plt.plot(K,silhouette_coeff, c ='r', marker = 'o')
plt.xlabel('Number of Cluster.')
plt.ylabel('Silhouette Co-efficient.')
```

```python
# Declaring Model
model = KMeans(n_clusters=3)
```

```python
model.fit(iris_df.data)
```

```python
# Predicting a single input
predict_label = model.predict([[7.2, 3.5, 0.8, 1.6]])
```

```python
# Predicion on entire data
all_predictions = model.predict(iris_df.data)
```

### DBSCAN clustering in Python

```python
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
```

```python
dbscan = DBSCAN()
```

```python
## Fitting
dbscan.fit(iris_df.data)
```

```python
# Transferi2ng using PCA
pca = PCA(n_components=2).fit(iris_df.data)
pca_2d = pca.transform(iris_df.data)
```

```python
dbscan.labels_
```

```python
## plot based on class
sns.set_style(style='white')
plt.title('DBSCAN finds 2 clusters and Noise')
for i in range(0, pca_2d.shape[0]):
    if dbscan.labels_[i] ==0:
        c1 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='r', marker='+', label = 'Cluster-1')
    elif dbscan.labels_[i] ==1:
        c2 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='g', marker='o', label = 'Cluster-2')
    elif dbscan.labels_[i] ==-1:
        c3 = plt.scatter(pca_2d[i,0], pca_2d[i, 1], c='b', marker='*', label = 'Cluster-3')
plt.legend(handles=[c1, c2, c3])
```
